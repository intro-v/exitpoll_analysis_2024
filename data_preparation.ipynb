```python
#function to check that no data get lost
def check_fun(df):
    print('# rows: ', df.shape[0])
    print('# УИКов: ', len(set(df['УИК'])))
    print('# Стран: ', len(set(df['Страна'])))
    print('# Городов: ', len(set(df['Город'])))

    print('# rows Аргентина: ', df[df['Страна'] == 'Аргентина / Argentina'].shape[0])
    print('# rows Веллингтон: ', df[df['Город'] == 'Веллингтон / Wellington'].shape[0])
    
    if 'count' in df.columns:
        print('# count all: ', df['count'].sum()) 
        print('# count Аргентина: ', df[df['Страна'] == 'Аргентина / Argentina']['count'].sum()) 
        print('# count Веллингтон: ', df[df['Город'] == 'Веллингтон / Wellington']['count'].sum())
```
